```python
#to merge the threee telcom tables
telcom_all = pd.concat((telcom_a_csv, telcom_b_csv, telcom_c_csv), axis=0, ignore_index=True)
telcom_all
```

```python
#checking the number of times telcom products were used on the 1st day
grouped = telcom_a_csv.groupby(['PRODUCT']) ['VALUE'].sum()
grouped
```

```python
#checking the number of times telcom products were used on the 2nd day
grouped1 = telcom_b_csv.groupby(['PRODUCT']) ['VALUE'].sum()
grouped1
```

```python
#checking the number of times telcom products were used on the 3rd day
grouped2 = telcom_c_csv.groupby(['PRODUCT']) ['VALUE'].sum()
grouped2
```

```python
#Cheking the total times products offered by telcom were used over the course of the three days
#the products were grouped and the sum of their values calculated
grouped3 = telcom_all.groupby(['PRODUCT']) ['VALUE'].sum()
grouped3
```
